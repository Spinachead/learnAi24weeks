```python
import numpy as np
import matplotlib.pyplot as plt

#y=3x+2
x_data = [1.0, 2.0, 3.0]
y_data = [5.0, 8.0, 11.0]

def forward(x, w, b):
    return x * w + b

def loss(x, y, w, b):
    y_pred = forward(x, w, b)
    return (y_pred - y) * (y_pred - y)

# 
W = np.arange(0.0, 4.1, 0.1)
B = np.arange(0.0, 4.1, 0.1)
w_mesh, b_mesh = np.meshgrid(W, B)

# 
l_sum = np.zeros_like(w_mesh)

# 
for x_val, y_val in zip(x_data, y_data):
    y_pred_val = forward(x_val, w_mesh, b_mesh)
    loss_val = (y_pred_val - y_val) ** 2
    l_sum += loss_val

l_sum = l_sum / len(x_data)

# 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(w_mesh, b_mesh, l_sum, cmap='viridis')
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('MSE')
plt.show()
```
